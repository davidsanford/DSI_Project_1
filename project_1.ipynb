##### 1. What does the data describe?

The data describes the SAT scores for student in all fifty states and the District of Columbia, along with a summary total.  It contains the participation rate and mean scores on the Math and Verbal portion of SAT I Reasoning Test.  The participation rate appears to refer to High School Seniors, according to a reading of the original data set documentation (https://research.collegeboard.org/programs/sat/data/archived/cb-seniors-2001).  The data is organized by default in terms of decreasing participation rate, with columns:

#### State, Rate, Verbal, Math

## Step 2: Load the data.

##### 4. Load the data into a list of lists

```python
import csv
sat_list = list(csv.reader(open("sat_scores.csv","r")))
```

##### 5. Print the data

```python
# Looping print statement used for better visualization over default print statement
for row in sat_list:
    print(row)
```

##### 6. Extract a list of the labels from the data, and remove them from the data.

```python
header = sat_list[0]
del sat_list[0]
```

##### 7. Create a list of State names extracted from the data. (Hint: use the list of labels to index on the State column)

```python
sat_states = []
state_idx = header.index('State')
del header[state_idx]

for row in sat_list:
    sat_states.append(row.pop(state_idx))
```

##### 8. Print the types of each column

```python
column_types = {}
for i, item in enumerate(header):
    column_types[header[i]] = type(sat_list[0][i])

print(column_types)
```

##### 9. Do any types need to be reassigned? If so, go ahead and do it.

```python
# All three columns are originally of 'str' type, but should be numeric for quantitative analysis.
sat_list = [[int(i) for i in row] for row in sat_list]

column_types = {}
for i, item in enumerate(header):
    column_types[header[i]] = type(sat_list[0][i])

print(column_types)
```

##### 10. Create a dictionary for each column mapping the State to its respective value for that column. 

```python
rate_dict = {}
verbal_dict = {}
math_dict = {}

for i, row in enumerate(sat_list):
    rate_dict[sat_states[i]] = row[header.index("Rate")]
    verbal_dict[sat_states[i]] = row[header.index("Verbal")]
    math_dict[sat_states[i]] = row[header.index("Math")]
    
print(rate_dict)
print(verbal_dict)
print(math_dict)
```

##### 11. Create a dictionary with the values for each of the numeric columns

```python
data_columns = {}

for i, colname in enumerate(header):
    data_columns[colname] = [row[i] for row in sat_list]

print(data_columns)
```

## Step 3: Describe the data

##### 12. Print the min and max of each column

```python
minmax = {"min" : {}, "max" : {}}

for colname, vals in data_columns.items():
    minmax["min"][colname] = None
    minmax["max"][colname] = None
    for num in vals:
        if minmax["min"][colname] is None or minmax["min"][colname] > num:
            minmax["min"][colname] = num
        if minmax["max"][colname] is None or minmax["max"][colname] < num:
            minmax["max"][colname] = num
            
print(minmax)
```

##### 13. Write a function using only list comprehensions, no loops, to compute Standard Deviation. Print the Standard Deviation of each numeric column.

```python
def stddev(column):
    mean = sum(column) / len(column)
    var = sum([(val - mean)**2 for val in column]) / len(column)
    return(var ** 0.5)

standard_deviations = {}
for colname, vals in data_columns.items():
    standard_deviations[colname] = stddev(vals)
    
print(standard_deviations)
```

##### 17. What is the typical assumption for data distribution?

The default assumption regarding arbitrary data is that it is reasonable to intially compare it to a gaussian distribution.

##### 18. Does that distribution hold true for our data?

The rate data is manifestly non-gaussian, possessing two distinct clusters with multi-modal behavior within the two clusters.  The Math data demonstrates multi-modal behavior, but may be marginally consistent with a normal distribution given the relatively large errors on the individual bins, as it is roughly symmetrical and possesses decreasing tails.  The Verbal score distribution appears more similar to a uniform distribution, as it is roughly constant within errors and possesses no apparent tails.

##### 19. Plot some scatterplots. **BONUS**: Use a PyPlot `figure` to present multiple plots at once.

```python
plt.figure(figsize=(10,5))

fig1 = plt.subplot(121)
plt.plot(data_columns["Math"], data_columns["Verbal"], 'g*',markersize=8,)

fig1.set_xlabel('Math Score',fontsize = 10)
fig1.set_ylabel('Verbal Score',fontsize = 10)
fig1.set_title('Verbal Scores vs. Math Scores',fontsize = 15)
fig1.set_xlim([400,650])
fig1.set_ylim([400,650])

fig2 = plt.subplot(122)
math_dots, = plt.plot(data_columns["Rate"], data_columns["Math"], 'bs',markersize=8)
verbal_dots, = plt.plot(data_columns["Rate"], data_columns["Verbal"], 'r^',markersize=8)
legend = plt.legend(handles = [math_dots,verbal_dots], labels = ["Math","Verbal"], title = "Test Type",fontsize = 10)
plt.setp(legend.get_title(),fontsize=15)

fig2.set_xlim([0,100])
fig2.set_ylim([400,650])

fig2.set_xlabel('Test Score',fontsize = 10)
fig2.set_ylabel('Senior Participation Rate',fontsize = 10)
fig2.set_title('Participation Rate vs. Test Scores',fontsize = 15)

plt.show()
```

##### 20. Are there any interesting relationships to note?

There is a positive correlation between the Verbal and Math test scores, with a slope close to one.  There is a negative correlation between both test scores and the participation rate, with similar clustering for the two tests, and the relationship has a non-linear component.

There is also a significant outlier point visible in both plots, and from the Rate vs. Scores plot its value for Math score (worst in the nation) is much more unusual given its participation rate than its Verbal score.  Examining the original data set, the outlier point corresponds to the state of Ohio.

In terms of interpretation of the data, the correspondence between Verbal and Math test scores is logical. It is reasonable to hypothesize that performance on tests should be correlated with study time, so on average who studies more will do better on both tests than a student who studies less, and the differences in their scores will average out over a population.

The rate might have a negative correlation with both scores because lower participation rate naturally suggests that only the most motivated students are taking the test, and those students are also the most likely to study.  As a result, the lower parts of the distribution are cut off by lower participation rate.
